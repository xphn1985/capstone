

Obtain data from the website



```python
req=requests.get(' https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')
soup = BeautifulSoup(req.content,'lxml')
table=soup.find_all('table')[0]
df=pd.read_html(str(table))
```

LOAD AND EXPLORE THE DATA

```python
neighbourhood = pd.DataFrame(df[0])
neighbourhood.head()
```


Removing the "not assigned" in Borough


```python
neighbourhood_filtered=neighbourhood[neighbourhood.Borough!="Not assigned"]
neighbourhood_filtered
```


JOIN THE ENTIRES WITH THE SAME POSTAL CODE


```python
neighbourhood_filtered.groupby("Postal Code")['Neighbourhood'].apply(','. join).reset_index()
```


CHECK IF THERE IS "NOT ASSIGNED" IN "NEIGHBOURHOOD"


```python
(neighbourhood_filtered["Neighbourhood"]=="Not assigned").sum()
```

Print the shape

```python
neighbourhood_filtered.shape
```


IMPORT THE LAT AND LONG DATA


```python
cord = pd.read_csv("Geospatial_Coordinates.csv")
cord.head()
```


COMBINING THE TWO DATAFRAMES


```python
neighbourhood_cord= pd.merge(neighbourhood_filtered, cord, on='Postal Code')
neighbourhood_cord.head()
```

SELECTING THE BOROUGHS CONTAINING  "TORONTO"

```python
toronto_data = neighbourhood_cord[neighbourhood_cord['Borough'].str.contains("Toronto")] .reset_index(drop=True)
toronto_data.head()
```

```python
address = 'Downtown Toronto'

geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Manhattan are {}, {}.'.format(latitude, longitude))
```


CREATING THE MAP


```python
# create map of Manhattan using latitude and longitude values
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=11)

# add markers to map
for lat, lng, label in zip(toronto_data['Latitude'], toronto_data['Longitude'], toronto_data['Neighbourhood']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
